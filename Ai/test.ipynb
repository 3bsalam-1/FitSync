```python
df.columns = df.columns.str.strip()
df['diabeties'] = df['Disease'].str.contains('diabeties')
df['heart_disease'] = df['Disease'].str.contains('heart_disease')
df['hypertension'] = df['Disease'].str.contains('hypertension')
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Sample categorical data

# Create an instance of LabelEncoder
encoder = LabelEncoder()

# Fit and transform the data
df['diabeties'] = encoder.fit_transform(df['diabeties'])
df['heart_disease'] = encoder.fit_transform(df['heart_disease'])
df['hypertension'] = encoder.fit_transform(df['hypertension'])
df
```

```python
df.drop('Disease', inplace=True, axis=1)
df.drop('Price', inplace=True, axis=1)

df
```
