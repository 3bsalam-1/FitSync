```python
import csv

# Sample data
healthy_person_data = [75, 76, 78, 80, 82, 79, 77, 76, 75, 74]
diabetic_data = [85, 88, 90, 92, 95, 91, 89, 88, 87, 86]
high_blood_pressure_data = [90, 92, 95, 98, 100, 96, 94, 92, 91, 90]
heart_patient_data = [100, 105, 110, 115, 120, 116, 112, 108, 105, 102]

# Combine data into a dictionary
data = {
    'Time': list(range(1, 11)),
    'Healthy Person': healthy_person_data,
    'Diabetic': diabetic_data,
    'High Blood Pressure': high_blood_pressure_data,
    'Heart Patient': heart_patient_data
}

# Write to CSV file
with open('heart_rate_data.csv', 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=data.keys())
    
    # Write header
    writer.writeheader()
    
    # Write data rows
    for i in range(10):
        writer.writerow({key: data[key][i] for key in data.keys()})
```

```python
import csv
import random

# Function to generate synthetic heart rate data
def generate_heart_rate_data():
    return [random.randint(60, 100) for _ in range(100)]

# Generate data for each category
healthy_person_data = generate_heart_rate_data()
diabetic_data = generate_heart_rate_data()
high_blood_pressure_data = generate_heart_rate_data()
heart_patient_data = generate_heart_rate_data()

# Combine data into a dictionary
data = {
    'Time': list(range(1, 101)),
    'Healthy Person': healthy_person_data,
    'Diabetic': diabetic_data,
    'High Blood Pressure': high_blood_pressure_data,
    'Heart Patient': heart_patient_data
}

# Write to CSV file
with open('heart_rate_big_data.csv', 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=data.keys())
    
    # Write header
    writer.writeheader()
    
    # Write data rows
    for i in range(100):
        writer.writerow({key: data[key][i] for key in data.keys()})
```

```python


df = pd.read_csv('H:\FitSync\Ai\data\heartbeat\heartbeats_data1.csv')
df[['min','max']] = df['Heartbeat Range'].str.split('-', expand=True)
df.drop(['Heartbeat Range'], axis=1, inplace=True)

df.to_csv('heartbeats2.csv', index=False)
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.capitalize()
```

```python
from sklearn.preprocessing import LabelEncoder

# Sample categorical data

# Create an instance of LabelEncoder
encoder = LabelEncoder()

# Fit and transform the data
df['Gender'] = encoder.fit_transform(df['Gender'])
df.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot scatter matrix to visualize relationships between numerical variables
sns.pairplot(df)
plt.show()
# Compute the correlation matrix
#correlation_matrix = df.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.boxplot(data=df, orient="h")
plt.xlabel("Value")
plt.ylabel("Frequency")
#plt.title("ID")
plt.tight_layout()
plt.show()
```

```python
for column in df.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[column], bins=30)
    plt.title(f'Distribution of {column}')
    plt.show()
```
