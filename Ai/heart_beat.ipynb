```python
# Function to detect outliers
def detect_outliers(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = dataframe[(dataframe[column] < Q1 - 1.5*IQR) | (dataframe[column] > Q3 + 1.5*IQR)]
    return outliers

# Detect and print number of outliers for each feature
for column in df.columns:
    outliers = detect_outliers(df, column)
    print(f'Number of outliers in {column}: {len(outliers)}')
```

```python
train.info()
```

```python
from sklearn.model_selection import train_test_split
x=train[train.columns.drop(['Min'],['Max'])]
y1=train[['Min']]
y2=train[['Max']]
x_train,x_val,y1_train,y1_val=train_test_split(x,y1,test_size=0.33,random_state=42)
x_train,x_val,y2_train,y2_val=train_test_split(x,y2,test_size=0.33,random_state=42)
```

```python
print('original tain shape : ',x.shape)

print('x_train shape :',x_train.shape)
print('x_val shape :',x_val.shape)

print('y1_train shape :',y1_train.shape)
print('y1_val shape :',y1_val.shape)
print('y2_train shape :',y2_train.shape)
print('y2_val shape :',y2_val.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_sc = pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns)
x_val_sc = pd.DataFrame(scaler.fit_transform(x_val), columns=x_val.columns)

```

```python
#linear regression model
#train data
model_1=LinearRegression()
model_2=LinearRegression()

model_1.fit(x_train_sc,y1_train)
model_2.fit(x_train_sc,y2_train)
```

```python
y1_predict=model_1.predict(x_val_sc)
y2_predict=model_2.predict(x_val_sc)
```
