```python
df.drop(['ID'], axis=1, inplace=True)
df.drop(['Stress'], axis=1, inplace=True)
df.drop(['Anxiety'], axis=1, inplace=True)
df.drop(['Alcohol'], axis=1, inplace=True)
df.drop(['Illness'], axis=1, inplace=True)
df.drop(['Hydration'], axis=1, inplace=True)
df.drop(['Other'], axis=1, inplace=True)
```

```python
# importing pandas package
import pandas as pd
 

 
# sorting dataframe
#df.sort_values("Team", inplace=True)
 
# making boolean series for a team name
filter1 = df["Age"] == 'Adult'
 
# making boolean series for age
filter2 = df["Age"] =='Older Adult'
 
# filtering data on basis of both filters
df.where(filter1 | filter2, inplace=True)
 
```

```python
df=df.dropna(how='any')
```

```python
from sklearn.preprocessing import LabelEncoder

# Sample categorical data

# Create an instance of LabelEncoder
encoder = LabelEncoder()

# Fit and transform the data
df['Gender'] = encoder.fit_transform(df['Gender'])
df['Age'] = encoder.fit_transform(df['Age'])
df['Medication'] = encoder.fit_transform(df['Medication'])
df['Caffeine'] = encoder.fit_transform(df['Caffeine'])
df['High_Pressure'] = encoder.fit_transform(df['High_Pressure'])
df['Low_Pressure'] = encoder.fit_transform(df['Low_Pressure'])
df['Heart_Condition'] = encoder.fit_transform(df['Heart_Condition'])
df['Diabetic'] = encoder.fit_transform(df['Diabetic'])





```
