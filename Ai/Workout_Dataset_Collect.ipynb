```python
import pandas as pd
import random
import os
```

```python
df=pd.read_csv('data/workout/workout_dataset.csv')
df
```

```python
df.columns = df.columns.str.strip()
df['Diabetes'] = df['Medical Condition'].str.contains('Diabetes')
df['Heart Disease'] = df['Medical Condition'].str.contains('Heart Disease')
df['Hypertension'] = df['Medical Condition'].str.contains('Hypertension')
df
```

```python
df.fillna(0, inplace=True)
df.replace(False, 0, inplace=True)
df.replace(True, 1, inplace=True)
df
```

```python
df.drop('Medical Condition', axis=1, inplace=True)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns
numerical_df = df.select_dtypes(include='number')

# Set the figure size
plt.figure(figsize=(10, 8))

# Plot correlation matrix
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')

```

```python
print("Descriptive Statistics:")
print(numerical_df.describe())

# Check for duplicates
print("Number of duplicate rows: ", numerical_df.duplicated().sum())

# Distribution plots for each feature
for column in numerical_df.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(numerical_df[column], bins=30, kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
import numpy as np

# Calculate z-scores for each numerical column
z_scores = (numerical_df - numerical_df.mean()) / numerical_df.std()

# Define a threshold for outlier detection (e.g., z-score > 3 or z-score < -3)
threshold = 3

# Find outliers by checking if the absolute z-score exceeds the threshold
outliers = np.abs(z_scores) > threshold

# Handle outliers by replacing them with NaN or a suitable value
numerical_df[outliers] = np.nan

# Alternatively, you can remove the rows containing outliers
# numerical_df = numerical_df[~outliers.any(axis=1)]

# Fill NaN values with a suitable method (e.g., mean, median, or interpolation)
numerical_df.fillna(numerical_df.mean(), inplace=True)

# Print the updated dataframe
numerical_df
```

```python
sns.boxplot(data=numerical_df, orient="h")
plt.xlabel("Value")
plt.ylabel("Frequency")
#plt.title("ID")
plt.tight_layout()
plt.show()
```
